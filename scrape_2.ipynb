Now that I have the basic workflow of scraping the data, I can start to scrape for multiple seasons/years of data. I'll basically put the workflow into a loop that starts from the 2014 season to current.

While thinking about scraping for college basketball data, I'm limiting the scope of the project once again by picking only one college conference, the Big 12. 

I also believe 2012 is a better point to start from because this was around the time the NBA began to evolve and attempt to shoot from beyond the arc. Starting from 2012, the increase in league average for 3 pointers attempted per season sky rocketed. (https://pmc.ncbi.nlm.nih.gov/articles/PMC9915101/#:~:text=In%201979%2C%20when%20the%20three,accessed%20on%201%20July%202019).)


```python
base_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"

# years to scrape
years = list(range(2003, 2024))  # 2003 to 2023

# columns to scrape
columns = ["Player", "Age", "Team", "Pos", 'PTS', 'AST', "TRB", 'ORB', 'DRB', 
'STL', 'BLK', 'TOV', "G", "MP", "FG", "FGA", "FG%", "eFG%", 'FT', 'FTA', 'FT%']

# empty DataFrame to hold all data
all_data = pd.DataFrame()
```

```python
# Loop through each year and scrape data
for year in years:
    url = base_url.format(year)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the table
    table = soup.find("table", {"id": "per_game_stats"})
    
    # read table into a dataframe
    df = pd.read_html(str(table))[0]
    
    
    # select columns 
    df = df[columns]
    
    # add a column for year
    df["Year"] = year
    
    # append to main DataFrame
    all_data = pd.concat([all_data, df])
    
    # rate limiting so I don't break this server
    time.sleep(1)

# Reset the index of the final DataFrame
all_data.reset_index(drop=True, inplace=True)


```

I don't really need the league average row so I can just drop that.

```python
season_avg_df = all_data[all_data['Player'] != 'League Average']
```

```python
new_column = ['Player', 'Age', 'Team', 'Pos', 'PTS', 'AST', 'TRB', 'ORB', 'DRB',
       'STL', 'BLK', 'TOV', 'G', 'MP', 'FG', 'FGA', 'FG%', 'eFG%', 'FT', 'FTA',
       'FT%', 'Year']
```

```python
new_column = [column.lower() for column in new_column]
```

```python
season_avg_df.columns = new_column
```

```python
season_avg_df['year'].unique()
```

```python
numeric_columns = ['pts', 'ast','trb', 'orb','drb', 'stl',
 'blk','tov','g','mp','fg','fga','fg%','efg%','ft','fta','ft%']
```

Instead of looking at everybody's season by season stats, I want to look at everybody's career averages. So I'll use groupby to get every player's averages on their statistics.

```python
career_avg_df = season_avg_df.groupby('player').agg(
    {column: 'mean' for column in numeric_columns}
).reset_index()
```

Making sure everybody's statistics are only 2 decimal places

```python
career_avg_df = career_avg_df.round(2)
```

Sorting values from greatest to least on points per game. 

Note LeBron at #4 at age 76

```python
career_avg_df = career_avg_df.sort_values(by = 'pts', ascending = False).reset_index(drop = True)
```

```python
sorted_season_avg_df = season_avg_df.sort_values(by = ['player', 'year'], ascending=True).reset_index(drop= True)
```
